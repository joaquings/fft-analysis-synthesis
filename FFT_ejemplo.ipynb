```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft,fftshift
from scipy.signal import windows
```

```python
def espectro(x):
  N = len(x)

  v = windows.triang(N)

  X = fft( x*v )

  Dw = 2*np.pi/N
  w = np.linspace( 0,2*np.pi-Dw, N)

  return w, np.abs(X)
```
